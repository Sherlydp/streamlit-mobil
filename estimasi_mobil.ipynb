# Nama mobil berdasarkan percepatan

```python
CarName  = df.groupby('CarName').count()[['horsepower']].sort_values(by='horsepower',ascending=True).reset_index()
CarName  = CarName .rename(columns={'horsepower':'numberOfCars'})
```

```python
fig = plt.figure(figsize=(15,5))
sns.barplot(x=CarName ['CarName'], y=CarName ['numberOfCars'], color='royalblue')
plt.xticks(rotation=60)
```

# Ukuran mesin

```python
engine = df.groupby('enginesize').count()[['horsepower']].sort_values(by='horsepower').reset_index()
engine = engine.rename(columns={'horsepower':'count'})
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x=engine['enginesize'], y=engine['count'], color='royalblue')
```

# Distribusi harga mobil

```python
plt.figure(figsize=(15,5))
sns.distplot(df['price'])
```

# Seleksi Fitur

```python
features = ['car_ID','symboling','wheelbase','carlength','carwidth','carheight','curbweight','enginesize','boreratio','stroke','compressionratio','horsepower','peakrpm','citympg','highwaympg']
x = df[features]
y = df['price']
x.shape, y.shape
```

# Split data training dan data seting

```python
from sklearn.model_selection import train_test_split
x_train, X_test, y_train, y_test = train_test_split(x,y,random_state=70)
y_test.shape
```

# Membuat model Regresi linear

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
pred = lr.predict(X_test)
```

```python
score = lr.score(X_test, y_test)
print('akurasi model regresi linier = ', score)
```

# Membuat inputan model regresi linear

```python
# car_ID=1, symboling=3, wheelbase=99.4, carlength=3.670, carwidth=1.695, carheight=1.605, curbweight=1.200, enginesize=152, boreratio=3.13, stroke=3.40, compressionratio=7.0, horsepower=160, peakrpm=4.250, citympg=19, highwaympg=26
input_data = np.array([[1,3,88.6,168.8,64.1,48.8,2548,130,3.47,2.68,9,111,5000,21,27]])

prediction = lr.predict(input_data)
print('Estimasi harga mobil dalam EUR :', prediction)
```
